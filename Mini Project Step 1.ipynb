```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler


df = pd.read_csv("retail_store_sales.csv")


print("First 5 rows:")
print(df.head())

#Display dataset information
print("\nDataset Info:")
print(df.info())

#Check missing values
print("\nMissing Values:")
print(df.isnull().sum())


# 1. Handle Missing Values


# Fill numerical missing values with median
num_cols = ['Price Per Unit', 'Quantity', 'Total Spent']

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

#Fill categorical missing values with mode
cat_cols = ['Category', 'Item', 'Payment Method', 'Location', 'Discount Applied']

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing values after handling:")
print(df.isnull().sum())


#2. Remove Duplicate Rows

df.drop_duplicates(inplace=True)

print("\nShape after removing duplicates:", df.shape)


#3. Convert Data Types

# Convert Transaction Date to datetime
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

# Convert Discount Applied to boolean
df['Discount Applied'] = df['Discount Applied'].astype(bool)

print("\nUpdated Data Types:")
print(df.dtypes)


#4. Encode Categorical Variables


label_encoder = LabelEncoder()

categorical_columns = ['Category', 'Item', 'Payment Method', 'Location']

for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

print("\nEncoded Dataset:")
print(df.head())


#5. Feature Scaling

scaler = StandardScaler()

scaled_columns = ['Price Per Unit', 'Quantity', 'Total Spent']

df[scaled_columns] = scaler.fit_transform(df[scaled_columns])

print("\nScaled Numerical Features:")
print(df.head())


#6. Outlier Detection

Q1 = df[scaled_columns].quantile(0.25)
Q3 = df[scaled_columns].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[scaled_columns] < (Q1 - 1.5 * IQR)) | 
            (df[scaled_columns] > (Q3 + 1.5 * IQR))).sum()

print("\nOutliers in each numerical column:")
print(outliers)


#7. Save Cleaned Dataset


df.to_csv("cleaned_retail_store_sales.csv", index=False)

print("\nData preprocessing completed successfully!")
print("Cleaned file saved as 'cleaned_retail_store_sales.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_retail_store_sales.csv")

# Set style
sns.set(style="whitegrid")

# -------------------------------
# 1. Basic Dataset Overview
# -------------------------------

print("Dataset Shape:")
print(df.shape)

print("\nFirst 5 Rows:")
print(df.head())

print("\nDataset Information:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())


#2. Check Missing Values


print("\nMissing Values:")
print(df.isnull().sum())


#3. Correlation Analysis

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()


#4. Distribution of Numerical Features


numerical_cols = ['Price Per Unit', 'Quantity', 'Total Spent']

for col in numerical_cols:
    plt.figure(figsize=(8,5))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()


#5. Boxplots for Outlier Detection

for col in numerical_cols:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()


#6. Sales by Category


plt.figure(figsize=(10,6))
sns.countplot(x='Category', data=df)
plt.title("Number of Sales by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()


#7. Sales by Payment Method


plt.figure(figsize=(10,6))
sns.countplot(x='Payment Method', data=df)
plt.title("Sales by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()


#8. Sales by Location

plt.figure(figsize=(10,6))
sns.countplot(x='Location', data=df)
plt.title("Sales by Location")
plt.xlabel("Location")
plt.ylabel("Count")
plt.show()

#9. Total Spent vs Quantity

plt.figure(figsize=(10,6))
sns.scatterplot(x='Quantity', y='Total Spent', data=df)
plt.title("Total Spent vs Quantity")
plt.show()

# -------------------------------
# 10. Pairplot
# -------------------------------

sns.pairplot(df[numerical_cols])
plt.show()

#11. Discount Applied Analysis

plt.figure(figsize=(8,5))
sns.countplot(x='Discount Applied', data=df)
plt.title("Discount Applied Distribution")
plt.show()

#12. Average Total Spent Analysis


print("\nAverage Total Spent:")
print(df['Total Spent'].mean())

print("\nMaximum Total Spent:")
print(df['Total Spent'].max())

print("\nMinimum Total Spent:")
print(df['Total Spent'].min())

eda_summary = df.describe()

eda_summary.to_csv("eda_summary.csv")

print("\nEDA completed successfully!")
print("EDA summary saved as 'eda_summary.csv'")
```
